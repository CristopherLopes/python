```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
#regression
x,y = make_regression(n_samples=200, n_features=1, noise=30)
plt.scatter(x,y)
```

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_

# y= a+b*x
# y=l_coeff+a_coeff*x
# y=-1.3608547638510622+55.19728849*x
```

```python
    print('Coeficiente Angular: {0}\nCoeficiente Linear {1}'.format(a_coeff,l_coeff))
```

```python
plt.scatter(x,y)
plt.plot(x, l_coeff + a_coeff*x, color='black')
plt.scatter(2.5, l_coeff + a_coeff*2.5, color='red', s= 130)
plt.show()
```

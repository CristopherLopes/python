```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
# gera dataset randômico
from sklearn.linear_model import LinearRegression
# executa a regressão linear
```

```python
# make regression, função
# informando parâmetros, 
# n_samples=200(número de amostras)
# n_features=1(número de categorias, no exemplo 1 por ser uma regressão simples )
# noise=30(ruído)
# Existe uma diferença entre o valor observado e o valor fornecido pela equação. 
# Esta diferença é denominada erro e é representada por ε, 
# é uma variável aleatória que quantifica a falha do modelo em ajustar-se aos dados exatamente.

x,y= make_regression(n_samples=200, n_features=1, noise=30)
# gera dataset randômico
plt.scatter(x,y)
# plotagem do gráfico
```

```python
modelo = LinearRegression()
modelo.fit(x,y)
```

```python
a_coeff = modelo.coef_
l_coeff = modelo.intercept_

# y= a+b*x
# y=mx+b ??? m->slope(declive inclinação)  b->y-intercept  
# y=l_coeff+a_coeff*x
# y=Coeficiente Linear+Coeficiente Angular*x
```

```python
    print('Coeficiente Angular: {0}\nCoeficiente Linear {1}'.format(a_coeff,l_coeff))
```

```python
plt.scatter(x,y)
plt.plot(x, l_coeff + a_coeff*x, color='black')
plt.scatter(2.5, l_coeff + a_coeff*2.5, color='red', s= 130)
plt.show()
```
